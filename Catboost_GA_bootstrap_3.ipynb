```python
# Load the data
data = pd.read_csv('/home/reza/Desktop/CatBoost/Hydraulic_conductivity/Book1.csv')
data
```

```python
# Split the data into features and target variable
X = data.drop('y', axis=1)
y = data['y']
```

```python
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
varbound = np.array([[10, 100],  # iterations range
                     [0.01, 0.3],  # learning rate range
                     [3, 8],  # depth range
                     [1, 5]])  # l2_leaf_reg range
vartype = np.array([['int'], ['real'], ['int'], ['real']])
```

```python
# 1. UTILITY FUNCTIONS
@contextmanager
def suppress_output():
    dummy_stdout = StringIO()
    dummy_stderr = StringIO()
    with redirect_stdout(dummy_stdout), redirect_stderr(dummy_stderr):
        yield

# 2. EVALUATION FUNCTIONS
def penalty_term(y_true, y_pred, threshold=0.001, base=3):
    residual = np.abs(y_true - y_pred)
    is_outside_threshold = residual > threshold
    residual_penalty = np.where(is_outside_threshold, np.exp(base * residual) - 1, 0)
    min_y_true = np.min(y_true)
    min_y_pred = np.min(y_pred)
    min_penalty = 0
    if min_y_pred < min_y_true:
        min_residual = min_y_true - min_y_pred
        min_penalty = np.exp(base * min_residual) - 1
    total_penalty = np.mean(residual_penalty) + min_penalty
    return total_penalty

def cv_fold(params, train_index, test_index, X, y):
    train_X, test_X = X.iloc[train_index], X.iloc[test_index]
    train_y, test_y = y.iloc[train_index], y.iloc[test_index]

    inner_cv = KFold(n_splits=5, shuffle=True, random_state=0)
    inner_mses = []

    for inner_train_index, inner_val_index in inner_cv.split(train_X):
        inner_train_X, inner_val_X = train_X.iloc[inner_train_index], train_X.iloc[inner_val_index]
        inner_train_y, inner_val_y = train_y.iloc[inner_train_index], train_y.iloc[inner_val_index]

        model = CatBoostRegressor(iterations=int(params[0]), learning_rate=params[1], depth=int(params[2]),
                                  l2_leaf_reg=params[3], random_seed=0, logging_level='Silent')
        model.fit(inner_train_X, inner_train_y, verbose=False)

        inner_y_pred = model.predict(inner_val_X)
        inner_mse = mean_squared_error(inner_val_y, inner_y_pred)
        inner_mses.append(inner_mse)

    model.fit(train_X, train_y, verbose=False)
    y_pred = model.predict(test_X)
    mse = mean_squared_error(test_y, y_pred)
    penalty = penalty_term(test_y, y_pred)
    
    return np.mean(inner_mses) + mse + penalty

def nested_cv_fold_bootstrap(params, X, y):
    outer_cv = KFold(n_splits=5, shuffle=True)
    inner_cv = KFold(n_splits=5, shuffle=True)
    outer_mses = []

    for outer_train_index, outer_test_index in outer_cv.split(X):
        outer_train_X, outer_test_X = X.iloc[outer_train_index], X.iloc[outer_test_index]
        outer_train_y, outer_test_y = y.iloc[outer_train_index], y.iloc[outer_test_index]
        
        inner_mses = []

        for inner_train_index, inner_test_index in inner_cv.split(outer_train_X):
            mse = cv_fold(params, inner_train_index, inner_test_index, outer_train_X, outer_train_y)
            inner_mses.append(mse)

        # Use the average inner MSE to represent the performance of the current outer fold
        outer_mses.append(np.mean(inner_mses))

    return np.mean(outer_mses)

def fitness_function_wrapper_bootstrap(X, y):
    def _fitness(params):
        outer_cv = KFold(n_splits=5, shuffle=True, random_state=0)
        outer_mses = [cv_fold(params, train_index, test_index, X, y) for train_index, test_index in outer_cv.split(X)]
        return np.mean(outer_mses)
    return _fitness

# MAIN BOOTSTRAP FUNCTION
def single_bootstrap_iteration(i):
    X_bootstrap, y_bootstrap = resample(X_train, y_train)
    X_bootstrap.iloc[:, 0] = np.clip(X_bootstrap.iloc[:, 0], None, 1.0)

    fit_func = fitness_function_wrapper_bootstrap(X_bootstrap, y_bootstrap)
    solver = ga(function=fit_func, dimension=4, variable_type_mixed=vartype, variable_boundaries=varbound,
                algorithm_parameters={'max_num_iteration': 20, 'population_size': 10, 'mutation_probability': 0.8,
                                      'elit_ratio': 0.02, 'crossover_probability': 0.9, 'parents_portion': 0.3,
                                      'selection_type': 'roulette', 'crossover_type': 'two_point',
                                      'mutation_type': 'random', 'max_iteration_without_improv': None}, function_timeout=150.0)
    solver.run()
    best_params = solver.output_dict['variable']
    model = CatBoostRegressor(iterations=int(best_params[0]), learning_rate=best_params[1], depth=int(best_params[2]), 
                              l2_leaf_reg=best_params[3], random_seed=0, logging_level='Silent')

    model.fit(X_bootstrap, y_bootstrap, verbose=False)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    penalty = penalty_term(y_test, y_pred)
    combined_mse = mse + penalty
    rmse = np.sqrt(combined_mse)
    r2 = r2_score(y_test, y_pred)

    return combined_mse, rmse, r2, best_params
    
# 4. EXECUTE BOOTSTRAP
warnings.filterwarnings('ignore')
with suppress_output():
    n_iterations = 800  # Adjust as required
    results = Parallel(n_jobs=-1, verbose=10)(delayed(single_bootstrap_iteration)(i) for i in range(n_iterations))

# 5. PROCESS BOOTSTRAP RESULTS
metric_distributions = {
    "mse": [],
    "rmse": [],
    "r2": [],
    "iterations": [],
    "learning_rate": [],
    "depth": [],
    "l2_leaf_reg": []
}
for mse, rmse, r2, best_params in results:
    metric_distributions["mse"].append(mse)
    metric_distributions["rmse"].append(rmse)
    metric_distributions["r2"].append(r2)
    metric_distributions["iterations"].append(best_params[0])
    metric_distributions["learning_rate"].append(best_params[1])
    metric_distributions["depth"].append(best_params[2])
    metric_distributions["l2_leaf_reg"].append(best_params[3])

results_df = pd.DataFrame(metric_distributions)
results_df.to_csv('bootstrap_results.csv', index=False)
```
